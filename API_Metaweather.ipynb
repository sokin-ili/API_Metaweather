# 1a. Get any 3 locations with location search

### Get data for the cities(Athens, Rome, London)

```python
cols = ['title', 'location_type', 'woeid', 'latt_long']
locations = pd.DataFrame(columns=cols)

cities = ['athens', 'rome', 'london']

for city in tqdm(cities):
    response=urlopen('https://www.metaweather.com/api/location/search/?query='+city)
    elevations=response.read()
    jsonResponse=json.loads(elevations.decode('utf-8'))
    for i in range(len(jsonResponse)):
        data = jsonResponse[i]
        locations = locations.append(data, ignore_index=True)
```

# 1b. Get 7 days forecast with location day 

### forecast for Athens

```python
cols=['title', 'id', 'weather_state_name', 'weather_state_abbr', 'wind_direction_compass', 'created', 'applicable_date', 'min_temp',
       'max_temp', 'the_temp', 'wind_speed', 'wind_direction', 'air_pressure', 'humidity', 'visibility', 'predictability']
forecast_athens=pd.DataFrame(columns=cols)

days = [str(x) for x in range(1,8)]

for day in tqdm(days):
    response=urlopen('https://www.metaweather.com/api/location/946738/2021/06/0'+day)
    elevations=response.read()
    jsonResponse=json.loads(elevations.decode('utf-8'))
    for i in range(len(jsonResponse)):
        data = jsonResponse[i]
        forecast_athens = forecast_athens.append(data, ignore_index=True)
        
forecast_athens['title'] = 'Athens'
forecast_athens
```

### forecast for Rome

```python
cols=['title', 'id', 'weather_state_name', 'weather_state_abbr', 'wind_direction_compass', 'created', 'applicable_date', 'min_temp',
       'max_temp', 'the_temp', 'wind_speed', 'wind_direction', 'air_pressure', 'humidity', 'visibility', 'predictability']
forecast_rome=pd.DataFrame(columns=cols)

days = [str(x) for x in range(1,8)]

for day in tqdm(days):
    response=urlopen('https://www.metaweather.com/api/location/721943/2021/06/0'+day)
    elevations=response.read()
    jsonResponse=json.loads(elevations.decode('utf-8'))
    for i in range(len(jsonResponse)):
        data = jsonResponse[i]
        forecast_rome = forecast_rome.append(data, ignore_index=True)
        
forecast_rome['title'] = 'Rome'
forecast_rome
```

### forecast for London

```python
cols = ['title', 'id', 'weather_state_name', 'weather_state_abbr', 'wind_direction_compass', 'created', 'applicable_date', 'min_temp',
       'max_temp', 'the_temp', 'wind_speed', 'wind_direction', 'air_pressure', 'humidity', 'visibility', 'predictability']
forecast_london = pd.DataFrame(columns=cols)

days = [str(x) for x in range(1,8)]

for day in tqdm(days):
    response = urlopen('https://www.metaweather.com/api/location/44418/2021/06/0'+day)
    elevations = response.read()
    jsonResponse = json.loads(elevations.decode('utf-8'))
    for i in range(len(jsonResponse)):
        data = jsonResponse[i]
        forecast_london = forecast_london.append(data, ignore_index=True)
        
forecast_london['title'] = 'London'
forecast_london
```

# 2. Store the data in a relational database

### Create table locations.db and store the locations data from dataframe locations

```python
# Create the database. I chose to create a database that is called: ‘locations.db‘
conn1 = sqlite3.connect('locations.db')
cursor = conn1.cursor()
```

```python
# Create the table
cursor.execute('CREATE TABLE loca(title text, location_type text, woeid text, latt_long text)')
conn1.commit()
```

```python
# Get from pandas DataFrame to SQL
locations.to_sql('loca', conn1, if_exists='replace', index = False)
```

```python
cursor.execute(
'''  
SELECT * 
FROM loca
'''
)
```

```python
# extract data from SQL table
for row in cursor.fetchall():
    print(row)
```

### Create table forecast_athens.db and store the forecast data from dataframe forecast_athens

```python
# Create the database. I chose to create a database that is called: ‘forecast_athens.db‘
conn2 = sqlite3.connect('forecast_athens.db')
cursor = conn2.cursor()
```
